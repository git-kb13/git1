```python
from sklearn.model_selection import train_test_split
target_y=data_set['PE']
feature_x=data_set.drop('PE',axis=1, inplace=False)
train_x,test_x,train_y,test_y=train_test_split(feature_x,target_y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
lr_model=LinearRegression()
lr_pl=Pipeline(steps=[('scalar',StandardScaler()),('lr',lr_model)])
lr_pl.fit(train_x,train_y)
print('co-efficient:',lr_model.coef_)
print('intercept:',lr_model.intercept_)
```

```python
from sklearn.metrics import r2_score
```

```python
predicted_y=lr_pl.predict(test_x)
r2_score(test_y,predicted_y)
```
